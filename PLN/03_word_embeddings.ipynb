## Objetivo
Entender la representación distribuida de palabras mediante Word2Vec y explorar relaciones semánticas en el espacio vectorial.

```python
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
import nltk
from nltk.corpus import brown
nltk.download('brown', quiet=True)
print("Librerías listas.")
```

```python
sentences = [simple_preprocess(' '.join(s)) for s in brown.sents()[:5000]]
print(f"Oraciones para entrenamiento: {len(sentences)}")
print(f"Ejemplo: {sentences[0][:10]}")
```

## 2. Entrenamiento del modelo Word2Vec

```python
model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=3,
    workers=4,
    epochs=10,
    sg=1  # Skip-gram
)
print(f"Vocabulario: {len(model.wv)} palabras")
print(f"Dimensión de los vectores: {model.wv.vector_size}")
```

## 3. Exploración semántica

```python
print("Palabras similares a 'man':")
for word, score in model.wv.most_similar('man', topn=5):
    print(f"  {word:<15} {score:.4f}")

print("\nPalabras similares a 'city':")
for word, score in model.wv.most_similar('city', topn=5):
    print(f"  {word:<15} {score:.4f}")
```

```python
sim = model.wv.similarity('king', 'queen')
print(f"Similitud king–queen: {sim:.4f}")
sim2 = model.wv.similarity('man', 'woman')
print(f"Similitud man–woman:  {sim2:.4f}")
```

## 4. Visualización PCA 2D

```python
words = ['man', 'woman', 'king', 'queen', 'city', 'country', 'dog', 'cat', 'car', 'bus']
words = [w for w in words if w in model.wv]
vectors = np.array([model.wv[w] for w in words])

pca = PCA(n_components=2)
coords = pca.fit_transform(vectors)

plt.figure(figsize=(8, 6))
plt.scatter(coords[:, 0], coords[:, 1], s=60, color='steelblue')
for i, word in enumerate(words):
    plt.annotate(word, (coords[i, 0]+0.01, coords[i, 1]+0.01), fontsize=11)
plt.title('Embeddings Word2Vec proyectados con PCA')
plt.xlabel('PC1'); plt.ylabel('PC2')
plt.grid(True, linestyle='--', alpha=0.4)
plt.savefig('word2vec_pca.png', dpi=120)
plt.show()
```

## Conclusiones
- Word2Vec aprende representaciones semánticas densas sin supervisión explícita.
- La geometría del espacio vectorial captura relaciones como género, jerarquía y analogías.
- PCA permite visualizar agrupaciones semánticas en 2D, confirmando que palabras de contextos similares se posicionan cerca.
