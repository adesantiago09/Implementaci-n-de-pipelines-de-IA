## Objetivo
Aplicar técnicas fundamentales de preprocesamiento de texto: tokenización, eliminación de stopwords, lematización y vectorización con TF-IDF.

```python
import nltk
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('punkt_tab', quiet=True)

print("Librerías cargadas correctamente.")
```

## 1. Corpus de ejemplo
Se utilizan reseñas de productos en español para demostrar el pipeline de preprocesamiento.

```python
corpus = [
    "El producto llegó en perfectas condiciones y muy rápido.",
    "No me gustó para nada, la calidad es muy mala y el precio es demasiado alto.",
    "Excelente relación calidad-precio, lo recomiendo ampliamente.",
    "El envío tardó mucho pero el producto está bien.",
    "Muy decepcionante, no cumple con lo que dice la descripción."
]

print(f"Documentos en el corpus: {len(corpus)}")
for i, doc in enumerate(corpus):
    print(f"  [{i}] {doc}")
```

## 2. Tokenización y limpieza

```python
def limpiar_texto(texto):
    texto = texto.lower()
    texto = re.sub(r'[^a-záéíóúüñ\s]', '', texto)
    tokens = word_tokenize(texto, language='spanish')
    return tokens

corpus_tokens = [limpiar_texto(doc) for doc in corpus]
for i, tokens in enumerate(corpus_tokens):
    print(f"Doc {i}: {tokens}")
```

## 3. Eliminación de stopwords

```python
stop_es = set(stopwords.words('spanish'))
corpus_filtrado = [
    [t for t in tokens if t not in stop_es]
    for tokens in corpus_tokens
]
for i, tokens in enumerate(corpus_filtrado):
    print(f"Doc {i}: {tokens}")
```

## 4. Lematización

```python
lemmatizer = WordNetLemmatizer()
corpus_lematizado = [
    [lemmatizer.lemmatize(t) for t in tokens]
    for tokens in corpus_filtrado
]
for i, tokens in enumerate(corpus_lematizado):
    print(f"Doc {i}: {tokens}")
```

## 5. Vectorización TF-IDF

```python
vectorizer = TfidfVectorizer(max_features=20)
X = vectorizer.fit_transform(corpus)
df_tfidf = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print("Matriz TF-IDF:")
print(df_tfidf.round(3).to_string())
```

## Conclusiones
- La tokenización descompone el texto en unidades mínimas (tokens).
- Las stopwords añaden ruido sin aportar significado semántico relevante.
- TF-IDF penaliza términos frecuentes en todos los documentos y premia los distintivos.
- Este pipeline es la base para tareas más complejas como clasificación o clustering de texto.
